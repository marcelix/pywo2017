<h3> Upis podataka u datoteku tipa .xls te njihov prikaz u konzoli. </h3>

```python
import xlwt
from random import randint
import xlrd
from tabulate import tabulate

jos = "da"
while (jos=="da"):

	rowNum = int(input("Redovi: "))
	colNum = int(input("Stupci: "))

	bookWrite = xlwt.Workbook()
	sheetWrite = bookWrite.add_sheet("Sheet1")

	#sadrzaj 
	gornjaGranica = 2
	for iteracija in range(0,colNum):
	    
	    sheetWrite.write(0,iteracija,"RandNum (1,"+str(gornjaGranica)+")") #zaglavlja

	    lista = []
	    for i in range(1,rowNum+1):
	        lista.append(randint(1,gornjaGranica))

	    for red in range(1,rowNum+1):
	        sheetWrite.write(red,iteracija,lista[red-1])
	        
	    gornjaGranica += 1

	bookWrite.save("randomNums.xls")

	#upis
	

	bookRead = xlrd.open_workbook("randomNums.xls")
	sheet = bookRead.sheet_by_index(0)

	vrijednostiZaglavlja = [i for i in range(0,colNum)]
	zaglavlja = []
    
    #isprobavanja ....
    #zapis podataka u rječnik -> svaki ključ ima vrijednosti jednog stupca ("lakša obrada")
    
	#sadrzaj = {}
	#for stupac in range(0,colNum):
	#    temp = []
	#    for red in range(0,rowNum+1):
	#        #ovo samo odvaja zaglavlja u posebnu listu
	#        if((stupac in vrijednostiZaglavlja) and red == 0):
	#           zaglavlja.append(sheet.cell_value(red,stupac))
	#        else:
	            #sadrzaj se sprema u privremenu listu
	#            temp.append(sheet.cell_value(red,stupac))
	    #privremena lista se dodaje rjecniku
	#    sadrzaj[stupac] = temp
	#            

    
	#zapis podataka u listu -> svaka podlista unutar sadrzajaZaIspis ima sadrzaj jednog reda tablice

	sadrzajZaIspis = []

	for red in range(0,rowNum+1):
	    temp = []
	    for stupac in range(0,colNum):
	        if((stupac in vrijednostiZaglavlja) and red == 0): 
	        	zaglavlja.append(sheet.cell_value(red,stupac))
	        else:  
	            temp.append(sheet.cell_value(red,stupac))
	    #privremena lista se dodaje rjecniku
	    prazno = len(temp)
	    if (prazno == 0): continue
	    else:
	        sadrzajZaIspis.append(temp)
	            

	print(tabulate(sadrzajZaIspis,headers=zaglavlja,tablefmt="fancy_grid")) #ispis tablice u konzoli
	print("\n")
	jos = (input("Izmjeni tablicu? (da/ne) ")).lower()
	print("\n")
```
