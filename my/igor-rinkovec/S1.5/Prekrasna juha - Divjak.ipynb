```python
import requests
from pprint import pprint
from bs4 import BeautifulSoup
from bs4 import NavigableString
```

```python
r = requests.get("http://www.jutarnji.hr/vijesti/hrvatska/skandal-u-ministarstvu-obrazovanja-voditeljica-kurikularne-reforme-pod-istragom-nova-ministrica-odmah-digla-uzbunu-zasto-nije-dosla-na-posao/6359926/")
```

```python
r.raise_for_status()
```

```python
soup = BeautifulSoup(r.content, "lxml")
```

```python
heading = soup.find("h1", attrs={"class":"title"}).getText().strip()
```

```python
content = {"heading": heading, "content":[]}
for tag in soup.find("section", attrs={"id":"CImaincontent"}).find("div"):
    if isinstance(tag, NavigableString):
        continue
    type = "unknown"
    if tag.name == "p":
        type = "paragraph"
    elif tag.name[0] == "h":
        type = "header"
    content["content"].append({"type": type, "content": tag.getText()})
pprint(content)
```

```python
from textwrap import wrap
from itertools import zip_longest

outputLeft = wrap("== " + content["heading"] + " ==", 30)
outputLeft = outputLeft + ["", ""]
for body in content["content"][0:len(content["content"])//2]:
    if body["type"] == "header":
        outputLeft = outputLeft + [""]
        outputLeft = outputLeft + wrap("\n\n= " + body["content"] + " =", 30)
    else:
        outputLeft = outputLeft + wrap(body["content"], 30)

outputRight = []
for body in content["content"][len(content["content"])//2:len(content["content"])]:
    if body["type"] == "header":
        outputRight = outputRight + [""]
        outputRight = outputRight + wrap("\n\n= " + body["content"] + " =", 30)
    else:
        outputRight = outputRight + wrap(body["content"], 30)
        
output = ["%-35s%-30s" % m for m in zip_longest(outputLeft, outputRight, fillvalue="")]

for l in output:
    print(l)
    
```
