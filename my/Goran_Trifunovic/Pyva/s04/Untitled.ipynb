```python

def f(x):
    def g(y):
        return x+y
    return g
```

```python
retfunc=f(1)
retfunc(6)
```

```python
def MakeGradingFunction(scale,grades):
    assert len(scale)+1==len(grades)
    def gf(x):
        for s,g in zip(scale,grades):
            if x<s:
                return g
        else:
            return grades[-1]
    return gf
fja=MakeGradingFunction((50,),grades=("fail","pass"))
fja(60)
```
